```python
#Step 3: Create a dataframe
#http://gokhanatil.com/2017/11/python-for-data-science-importing-xml-to-pandas-dataframe.html
dfcols = ['cve', 'av', 'baseScore','ac','gdt']
df_xml02 = pd.DataFrame(columns=dfcols)
df_xml03 = pd.DataFrame(columns=dfcols)
df_xml04 = pd.DataFrame(columns=dfcols)
df_xml05 = pd.DataFrame(columns=dfcols)
df_xml06 = pd.DataFrame(columns=dfcols)
df_xml07 = pd.DataFrame(columns=dfcols)
df_xml08 = pd.DataFrame(columns=dfcols)
df_xml09 = pd.DataFrame(columns=dfcols)
df_xml10 = pd.DataFrame(columns=dfcols)
df_xml11 = pd.DataFrame(columns=dfcols)
df_xml12 = pd.DataFrame(columns=dfcols)
df_xml13 = pd.DataFrame(columns=dfcols)
df_xml14 = pd.DataFrame(columns=dfcols)
df_xml15 = pd.DataFrame(columns=dfcols)
df_xml16 = pd.DataFrame(columns=dfcols)
df_xml17 = pd.DataFrame(columns=dfcols)
df_xml18 = pd.DataFrame(columns=dfcols)
df_xmlmd = pd.DataFrame(columns=dfcols)
```

```python
#Step 4: Define the namespace
cvss = {'cvss':'http://scap.nist.gov/schema/cvss-v2/0.2'}
namespace = {"namespace":"http://scap.nist.gov/schema/feed/vulnerability/2.0"}
```

```python
#Step 5: Loop through to get data for each year, 2002
for entry in root02.findall('.//namespace:entry',namespace):
    entry_id = entry.get('id')
    for av in entry.findall('.//cvss:access-vector', cvss):
        av = av.text
    for score in entry.findall('.//cvss:score', cvss):
        score = score.text
    for auth in entry.findall('.//cvss:authentication', cvss):
        auth = auth.text
    for ac in entry.findall('.//cvss:access-complexity', cvss):
        ac = ac.text
    for gdt in entry.findall('.//cvss:generated-on-datetime', cvss):
        gdt = gdt.text
    df_xml02 = df_xml02.append(
        pd.Series([entry_id,av,score,ac,gdt], index=dfcols),ignore_index=True)
```
